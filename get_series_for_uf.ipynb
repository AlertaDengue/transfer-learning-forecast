```python
def get_curve_state(state, sigla):
    
    df_names = pd.read_sql_query(f'select geocodigo, nome, uf from "Dengue_global"."Municipio" WHERE uf = \'{state}\';', conexao)

    
    df_uf = pd.DataFrame()
    for geocodigo in df_names.geocodigo:
        df_cases = pd.read_sql_query(f'select "data_iniSE", casos from "Municipio"."Historico_alerta" where municipio_geocodigo={geocodigo} order by "data_iniSE" ASC', conexao)
        df_cases.set_index('data_iniSE', inplace = True)
        df_cases.index = pd.to_datetime(df_cases.index)

        df_uf[f'casos_{geocodigo}'] = df_cases.casos
        

    df_uf.to_csv(f'./data/casos_{sigla}.csv', index_label = 'data_iniSE')
```

```python
%%time

for n, s in zip(['Pernambuco', 'Paraíba', 'Paraná'], ['PE', 'PB', 'PR']):

    print(n)
    print(s)
    get_curve_state(n, s)
```

```python
df_uf = pd.read_csv('/Users/eduardoaraujo/Downloads/memUF2023_corrigido.csv')
del df_uf['Unnamed: 0']
df_uf = df_uf.rename(columns = {'V1': 'Low', 'V2': 'Median', 'V3': 'High'})

df_uf.head()
```

```python
df_uf.UF.unique()
```

```python
name_to_abbrev = {'Acre':'AC', 'Rondônia':'RO', 'Roraima':'RR', 'Amazonas':'AM', 'Pará':'PA', 'Amapá':'AP',
       'Tocantins':'TO', 'Mato Grosso':'MT', 'Mato Grosso do Sul':'MS', 'Goiás':'GO',
       'Distrito Federal':'DF', 'Rio Grande do Sul':'RS', 'Santa Catarina':'SC',
       'Paraná':'PR', 'Rio de Janeiro':'RJ', 'Minas Gerais':'MG', 'Espírito Santo':'ES',
       'São Paulo':'SP', 'Alagoas':'AL', 'Bahia':'BA', 'Ceará':'CE', 'Maranhão':'MA', 'Paraíba':'PB',
       'Pernambuco':'PE', 'Piauí':'PI', 'Rio Grande do Norte':'RN', 'Sergipe':'SE'}
```

```python
df_uf['sigla'] = df_uf.UF.replace(name_to_abbrev)

df_uf
```

```python
df_uf.isnull().sum()
```

```python
pop_uf = pd.read_csv('pop_uf.csv')

pop_uf
```

```python
df_uf.shape
```

```python
df_uf_pop = df_uf.merge(pop_uf, left_on = 'sigla', right_on = 'state')

df_uf_pop
```

```python
df_uf_pop.drop(['state'], axis = 1, inplace = True)

df_uf_pop = df_uf_pop.rename(columns = {'sigla': 'UF_id'})


df_uf_pop
```

```python
df_uf_pop['LowCases'] = df_uf_pop['Low']*df_uf_pop['pop']/(100000)
df_uf_pop['MedianCases'] = df_uf_pop['Median']*df_uf_pop['pop']/(100000)
df_uf_pop['HighCases'] = df_uf_pop['High']*df_uf_pop['pop']/(100000)

df_uf_pop
```

```python
df_uf_pop.to_csv('typical_inc_curves_uf.csv', index = False) 
```

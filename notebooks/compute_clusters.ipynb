```python
import sys
import pickle
import numpy as np
import pandas as pd
sys.path.append('../')
from clustering import compute_clusters, lag_ccf
```

# Ceará

```python
df = pd.read_csv('../data/casos_CE.csv')
df.set_index('data_iniSE', inplace = True)
df.index = pd.to_datetime(df.index)
df = df.fillna(0)

for col in df.columns: 
    
    if sum(df[col]) <= 50:
        
        df.drop(col, axis =1, inplace = True)
        
    else:
        df.rename(columns = {col:col[-7:]}, inplace = True)
    
df
```

```python
clusters, fig = compute_clusters(df, lags = 7, t = 2, plot = True)
```

```python
state = 'CE'

with open('../clusters/clusters_{}.pkl'.format(state), 'wb') as fp:
        pickle.dump(clusters, fp)
        
clusters
```
